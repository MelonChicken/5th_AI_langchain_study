```python
import os
from langchain_experimental.text_splitter import SemanticChunker
from langchain_google_genai import GoogleGenerativeAIEmbeddings

# êµ¬ê¸€ API í‚¤ ì„¤ì • (ì½”ë©ì˜ ê²½ìš° userdata ë˜ëŠ” í™˜ê²½ë³€ìˆ˜ ì‚¬ìš©)
os.environ["GOOGLE_API_KEY"] = "AIzaSyA-KasOCjOpTSq-mH_Pke-O_fj82_idAaM"

# Gemini ì„ë² ë”© ëª¨ë¸ ì´ˆê¸°í™” (ê°€ì¥ ìµœì‹ ì¸ models/embedding-001 ì‚¬ìš©)
embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001")

# ë°ì´í„° ë¡œë“œ (ì´ì „ ë‹¨ê³„ì—ì„œ ìƒì„±í•œ íŒŒì¼)
with open("./data/appendix-keywords.txt", "r", encoding="utf-8") as f:
    file = f.read()

# 1. ê¸°ë³¸ ì„¤ì • (Percentile ê¸°ë°˜)
text_splitter = SemanticChunker(
    embeddings,
    breakpoint_threshold_type="percentile",
    breakpoint_threshold_amount=70
)

# í…ìŠ¤íŠ¸ ë¶„í•  ë° ë¬¸ì„œ ìƒì„±
docs = text_splitter.create_documents([file])

print(f"ìƒì„±ëœ ì²­í¬ ìˆ˜: {len(docs)}")
print(f"ì²« ë²ˆì§¸ ì²­í¬ ë‚´ìš©:\n{docs[0].page_content}")
```
